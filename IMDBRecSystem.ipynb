```python
import pandas as pd
from sklearn.metrics import pairwise_distances
links = pd.read_csv('/content/links.csv')
movies = pd.read_csv('/content/movies.csv')
ratings = pd.read_csv('/content/ratings.csv')
tags = pd.read_csv('/content/tags.csv')
```

```python
print(links.head())
print(movies.head())
print(ratings.head())
print(tags.head())
```

```python
df = pd.merge(movies,ratings,on='movieId')
df
```

```python
df = df[['userId','title','rating']]
```

```python
pivot = df.pivot_table(index='title',columns='userId',values='rating',fill_value=0)
pivot
```

```python
corr_matrix = pairwise_distances(pivot,metric='correlation')
corr_matrix
```

```python
indexes = list(pivot.index)
movie_index = indexes.index('Godfather, The (1972)')
movie_index
```

```python
like_indexes =(corr_matrix[movie_index].argsort()[1:6])
like_indexes
```

```python
for i in like_indexes:
  print(indexes[i])
```

```python
like_indexes =(corr_matrix[indexes.index('Toy Story (1995)')].argsort()[1:10])
like_indexes
for i in like_indexes:
  print(indexes[i])
```
